## Seaborn Basic
### Questions solved from this website https://www.kaggle.com/code/swapnilkumarrai/seaborn-exercises/edit

```python
import seaborn as sns
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.sample(5)
type(titanic[(titanic['sex']=='male') & (titanic['class']=='Third')].count())
```

```python
sns.jointplot(data=titanic, x='fare', y='age')
```

```python
sns.displot(data=titanic['fare'], bins=30, kde=False, color='red')
```

```python
sns.boxplot(data=titanic, x='class', y='age', palette='rainbow')
```

```python
sns.swarmplot(data=titanic, x='class', y='age', palette='rainbow')
```

```python
sns.countplot(x='sex', data=titanic)
```

```python
# sns.heatmap(data=titanic.corr(),cmap='coolwarm')
# plt.title('titanic.corr()')
# Heatmap

# Plot rectangular data as a color-encoded matrix
temp_df = titanic.pivot(columns='sex',values='survived')

# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df, cmap='coolwarm')
```
